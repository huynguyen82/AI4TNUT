### 1.1.2.Các khái niệm và thuật ngữ (*Nguồn chatGPT, NVIDIA*)

Deep_Learning_Icons_R5_PNG.jpg.png
1. Trí tuệ nhân tạo - AI:

  *Trí tuệ nhân tạo (Artificial Intelligence - AI) là một lĩnh vực của khoa học máy tính tập trung vào việc phát triển các hệ thống và chương trình máy tính có khả năng thực hiện những tác vụ mà thường đòi hỏi trí thông minh của con người. Điều này bao gồm khả năng học hỏi, lập luận, giải quyết vấn đề, hiểu ngôn ngữ tự nhiên, nhận diện, và ra quyết định.*
2. Học máy – Machine learning

  *Là một nhánh của AI, là các thuật toán được phát triển để máy tính có thể học từ dữ liệu và cải thiện hiệu suất theo thời gian mà không cần được lập trình cụ thể cho từng tác vụ.*
3. Học sâu – Deep learning

  *Là một nhánh con của Machine Learning (Học máy), sử dụng các mạng nơ-ron sâu để xử lý và phân tích dữ liệu. Khác với các phương pháp học máy truyền thống, học sâu có khả năng tự động rút ra đặc trưng từ dữ liệu.*
4. AI tạo sinh
  *Generative AI là một lĩnh vực trong trí tuệ nhân tạo tập trung vào việc tạo ra nội dung mới, bao gồm văn bản, hình ảnh, âm thanh, và video, dựa trên các mẫu hoặc dữ liệu đã học. Thay vì chỉ phân tích hoặc nhận diện dữ liệu, Generative AI có khả năng "sáng tạo" nội dung mới, mang lại nhiều ứng dụng trong các lĩnh vực khác nhau.*
5. Bigdata:
  *BigData (Dữ liệu lớn) là thuật ngữ chỉ các tập dữ liệu cực kỳ lớn, phức tạp, đa dạng và tăng trưởng nhanh, khiến cho các phương pháp xử lý dữ liệu truyền thống không đủ khả năng để xử lý và phân tích chúng một cách hiệu quả. Những tập dữ liệu này thường đòi hỏi các công cụ và công nghệ tiên tiến để thu thập, lưu trữ, quản lý và phân tích nhằm trích xuất các thông tin có giá trị.  Các đặc trưng chính của Big Data (thường được gọi là "3 Vs"):*

    * Volume (Khối lượng): Khối lượng dữ liệu khổng lồ được tạo ra mỗi giây, thường đo bằng terabyte hoặc petabyte. \\
    * Velocity (Tốc độ): Tốc độ mà dữ liệu mới được tạo ra và di chuyển. Điều này bao gồm dữ liệu thời gian thực hoặc luồng dữ liệu. \\
    * Variety (Đa dạng): Các loại dữ liệu khác nhau, bao gồm dữ liệu có cấu trúc (như cơ sở dữ liệu), dữ liệu phi cấu trúc (như văn bản, hình ảnh, video), và dữ liệu bán cấu trúc (như các tệp XML).*

    > Dữ liệu có cấu trúc là loại dữ liệu được tổ chức theo một định dạng hoặc cấu trúc nhất định, thường là trong các bảng, cơ sở dữ liệu, hoặc các tệp tin có cấu trúc như CSV. Dữ liệu này dễ dàng tìm kiếm, quản lý và xử lý bằng các công cụ truyền thống như SQL: Dữ liệu trong cơ sở dữ liệu quan hệ như MySQL, Oracle; Bảng tính Excel chứa các hàng và cột có tiêu đề; Dữ liệu từ các biểu mẫu với các trường thông tin như tên, tuổi, địa chỉ. \\

    >Dữ liệu không có cấu trúc là loại dữ liệu không tuân theo một mô hình hoặc cấu trúc xác định. Dữ liệu này có thể ở nhiều dạng khác nhau và khó khăn hơn trong việc lưu trữ, tìm kiếm và xử lý. Phân tích dữ liệu không có cấu trúc thường đòi hỏi các công cụ và kỹ thuật phức tạp như xử lý ngôn ngữ tự nhiên (NLP) hoặc học máy (machine learning): Văn bản tự do như email, tài liệu Word; Hình ảnh, video, âm thanh; Các bài đăng trên mạng xã hội, blog.`
